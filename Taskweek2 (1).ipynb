```python
import pandas as pd

```

```python
data = {

    "Name": ["Ali", "Maria", "John", "Sara", "David", "Emma"],

    "Age": [25, 32, 45, 29, 40, 36],

    "Department": ["HR", "Finance", "IT", "HR", "IT", "Finance"],

    "Salary": [3000, 4000, 5000, 3200, 5500, 4200]

}
```

```python
df = pd.DataFrame(data)

print(df)
```

```python
print(df.head(3))
```

```python
print(df.info)
```

```python
print(df.describe)
```

```python

bins = [0, 25, 45, float('inf')]

labels = ['young', 'middle', 'senior']

df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

df
```

```python
df['Age_group'].count()
```

```python

age_group_counts = df['Age_group'].value_counts()


print("Number of employees in each age group:")
print(age_group_counts)
```

```python
average_salary_per_department = df.groupby('Department')['Salary'].mean()

print("Average Salary per Department:")
print(average_salary_per_department)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart for average salary per department
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_per_department.index, y=average_salary_per_department.values, palette='viridis')
plt.title('Average Salary per Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create a histogram of employee ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=5, kde=True, palette='coolwarm') # Using 5 bins for this small dataset, kde=True adds a density curve
plt.title('Distribution of Employee Ages')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
