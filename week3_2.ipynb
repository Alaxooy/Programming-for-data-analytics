```python
df['director'].fillna('Unknown', inplace=True)

df['cast'].fillna('Unknown', inplace=True)

df['country'].fillna('Unknown', inplace=True)
```

```python
#Task 2
```

```python
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
duplicates_removed = initial_rows - df.shape[0]
```

```python
df.dropna(subset=['rating'], inplace=True)

df['date_added'] = df['date_added'].str.strip() # Strip whitespace
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y') # Specify format
df['year_added'] = df['date_added'].dt.year

display(df.head())
```

```python
if 'duration' in df.columns:
    df['duration_minutes'] = df['duration'].apply(lambda x: int(x.split(' ')[0]) if 'min' in x else 0)
    df['duration_seasons'] = df['duration'].apply(lambda x: int(x.split(' ')[0]) if 'Season' in x else 0)
    df.drop(columns=['duration'], inplace=True)
    print("Duration columns created and original 'duration' column dropped.")
else:
    print("The 'duration' column does not exist. Assuming duration processing has already occurred.")

display(df.head())
```

### Encoding categorical column `type`

The `type` column indicates whether a title is a 'Movie' or a 'TV Show'. I will use one-hot encoding to convert this categorical column into numerical format, which is suitable for machine learning models.

```python
df = pd.get_dummies(df, columns=['type'], drop_first=True, dtype=int)

display(df.head())
```

### Creating new features: Content Age and Number of Genres

I will create two new features:
- **Content Age**: Calculated as the current year (2024) minus the `release_year`.
- **Number of Genres**: Derived by counting the number of genres listed in the `listed_in` column for each title.

```python
df['content_age'] = 2024 - df['release_year']
df['num_genres'] = df['listed_in'].apply(lambda x: len(x.split(', ')))

display(df.head())
```
