```python
import pandas as pd
import numpy as np
```

```python
product_prices = np.array([[100, 200, 150, 300],
[120, 180, 160, 280],
[90, 220, 140, 310],
[110, 210, 170, 290]])

tax_prices = np.array([0.05, 0.10, 0.08, 0.12])
print(tax_prices)
```

```python
applied_rates =  product_prices * (1 + tax_prices)
print(applied_rates)
```

```python
new_rates = np.sum(applied_rates,axis=0)
```

```python
highest_total = np.argmax(new_rates)
```
