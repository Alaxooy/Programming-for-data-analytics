## Section 1: Create and write to a text file
We will create a text file and write some lines into it.

```python
import os
import csv
import numpy as np

text_path = "arden_notes.txt"

with open(text_path, "w", encoding="utf-8") as f:
    f.write("Hello Arden University!\n")
    f.write("File handling in Python is simple.\n")
    f.write("strip and split help clean and parse lines.\n")

print("File created:", os.path.exists(text_path))
```

## Section 2: Read the full file
Now we will read the complete content of the file.

```python
with open(text_path, "r", encoding="utf-8") as f:
    content = f.read()

print(content)
```

## Section 3: Read part of the file
Now we will read only a few characters from the file.

```python
with open(text_path, "r", encoding="utf-8") as f:
    first_10 = f.read(10)
    next_8 = f.read(8)

print(first_10)
print(next_8)
```

## Section 4: Read line by line
Now we will read each line and split it into words.

```python
with open(text_path, "r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        words = line.strip().split()
        print(f"Line {i}:", words)
```

## Section 5: Append new content
Now we will add one more line to the file.

```python
with open(text_path, "a", encoding="utf-8") as f:
    f.write("Appending a final line for the demo.\n")

with open(text_path, "r", encoding="utf-8") as f:
    print(f.read())
```

## Section 6: Create a CSV file
Now we will create a simple CSV file with some data.

```python
csv_path = "employees.csv"

rows = [
    ["name", "employee_no", "school"],
    ["Angela Armstrong", "006001", "Chemical"],
    ["Stuart Smith", "006002", "Structural"]
]

with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(rows)

print("CSV created:", os.path.exists(csv_path))
```

## Section 7: Read the CSV file
Now we will read and display the CSV data.

```python
with open(csv_path, "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for i, row in enumerate(reader):
        if i == 0:
            print("Columns:", row)
        else:
            print(row)
```

## Section 13: Indexing and slicing
Now we will select specific parts of an array.

```python
a = np.array([
    [-1, 2, 0, 4],
    [4, -0.5, 6, 0],
    [2.6, 0, 7, 8],
    [3, -7, 4, 2.0]
])

print(a[:2, ::3])
```

## Section 14: Integer indexing
Now we will pick specific values using positions.

```python
print(a[[1, 1, 0, 3], [3, 2, 1, 0]])
```

## Section 15: Boolean masking
Now we will filter data using conditions.

```python
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data = np.array([
    [4, 7],
    [0, 2],
    [-5, 6],
    [0, 0],
    [1, 2],
    [-12, -4],
    [3, 4]
])

mask = names == "Bob"

print(data[mask])
print(data[~mask])
```

## Section 19: Aggregations
Now we will find max, min, and sum.

```python
u = np.array([
    [1, 5, 6],
    [4, 7, 2],
    [3, 1, 9]
])

print(u.max())
print(u.min())
print(u.sum())
```

## Section 20: Matrix operations
Now we will compare elementwise and matrix multiplication.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[4, 3], [2, 1]])

print(A * B)
print(A.dot(B))
```
