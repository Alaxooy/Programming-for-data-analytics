```python
import string

punctuations = string.punctuation
translator = str.maketrans('', '', punctuations)

all_words = []

with open(text_path, "r", encoding="utf-8") as f:
   for i, line in enumerate(f, start=1):

      clean_line = line.lower().translate(translator).strip()

      words = clean_line.split()


      all_words.extend(words)

print("All words:", all_words)
```

```python
from collections import Counter

total_words = len(all_words)
print(f"Total number of words: {total_words}")


unique_words = len(set(all_words))
print(f"Number of unique words: {unique_words}")

if all_words:
    word_counts = Counter(all_words)
    most_common_word = word_counts.most_common(1)[0][0]
    print(f"Most frequent word: {most_common_word}")
else:
    print("No words found to determine the most frequent.")

class_python_count = 0
for word in all_words:
    if word == 'class' or word == 'python':
        class_python_count += 1
print(f"Count of 'class' or 'python': {class_python_count}")
```
