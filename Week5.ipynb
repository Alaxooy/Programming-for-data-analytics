```python
import numpy as np
```

```python
energy = np.array([100, 120, 130, 150, 170, 160, 110])
print(energy)
```

```python
# Calculate indices for the middle 50% of elements using slicing
n = len(energy)

# Determine the number of elements to select (approx 50%)
no_of_elements = int(round(n * 0.5))

# Calculate start and end indices to center the slice
start = (n - no_of_elements) // 2
end = start + no_of_elements

# Extract the middle data using slicing
middle = energy[start:end]

print(f"Original array: {energy}")
print(f"Length of array: {n}")
print(f"Number of elements selected (approx 50%): {no_of_elements}")
print(f"Start index: {start}, End index: {end}")
print(f"Extracted middle data (using slicing): {middle}")
```

```python
# To ensure the original array is not modified, use .copy()
middle_copied = energy[start:end].copy()

print(f"Extracted middle data (using slicing with copy()): {middle_copied}")

# Modify the copied middle data to demonstrate non-modification of the original
middle_copied[0] = 999

print(f"\nModified copied middle data: {middle_copied}")
print(f"Original array after modifying copied data: {energy}")

# The original 'energy' array remains unchanged, confirming that .copy() created an independent array.
```
