```python
import pandas as pd
url = "/content/netflix_titles.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


type_counts = df['type'].value_counts()


plt.figure(figsize=(6,7))
sns.barplot(x=type_counts.index, y=type_counts.values)
plt.title('Distribution of Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('No of Titles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
countries_df = df.dropna(subset=['country'])

country_counts = countries_df['country'].str.split(', ', expand=True).stack()

top_5_countries = country_counts.value_counts().head(5)
display(top_5_countries)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_countries.index, y=top_5_countries.values, palette='plasma')
plt.title('Top 5 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
release_year_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(15, 7))
sns.barplot(x=release_year_counts.index, y=release_year_counts.values, palette='viridis')
plt.title('Number of Netflix Releases by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
most_releases_year = release_year_counts.idxmax()
max_releases_count = release_year_counts.max()
print(f"The year with the most Netflix releases is {most_releases_year} with {max_releases_count} titles.")
```
